## Teorema do Limite Central (TLC)

Seja X uma variável aleatória de interesse com distribuição qualquer, cuja média populacional e variância populacional são, usualmente, denotadas por E(X) = μ e Var(X) = σ2.  
Para estimar a média populacional, μ, utiliza-se o estimador Média Amostral definido por: 
X ̅  = ∑ Xi n i=1 n

## Item 2

##### Distribuição da Média Amostral para n = 2

```python
n = 2

rc = np.random.choice(x,n)
print(rc)
```

##### Criação da função que calcula a média

```python
def media(rc):
    soma= 0
    for i in range(len(rc)):
        soma += rc[i]
    media = soma/len(rc)
    return media

f = media(rc)
print(f)
```

##### Criação da função que gera a lista das médias

```python
def func_m(x,n):
    lista = []
    for i in range(10000):
        rc = np.random.choice(x,n)
        f = media(rc)
        lista.append(f)
    return lista

func = func_m(x,n)
print(func)
```

##### Tornando a lista das médias em um Dataframe

```python
h = pd.Series(data=sorted(func))
```

## Item 5

##### Criação das listas X1 e X9

```python
def X(x,n3):
    X1 = []
    X9 = []
    for i in range(10000):
        rc = np.random.choice(x,n3)
        X1.append(rc[0])
        X9.append(rc[8])
    return (X1, X9)

X19 = X(x,n3)

X1 = X19[0]
X9 = X19[1]
```

##### Histograma e distribuição de X1 

```python
h4 = pd.Series(data=sorted(X1))
faixa4 = np.arange(0,(h4.max()+1),1)
plot = h4.plot.hist(bins=faixa4, normed=True)
plt.plot(h4, stats.poisson.pmf(5, h4), color="r")
```

##### Histograma e distribuição de X9

```python
h5 = pd.Series(data=sorted(X9))
faixa5 = np.arange(0,(h5.max()+1),1)
plot = h5.plot.hist(bins=faixa5, normed=True)
plt.plot(h5, stats.poisson.pmf(5, h5), color="r")
```

Como pode ser visualizado as distribuições criadas nos dois histogramas acima podem ser comparadas à distribuição teórica de Poisson gerada no Item 1, uma vez que todas elas se assemelham muito, ou seja, suas distribuições são muito semelhantes.
